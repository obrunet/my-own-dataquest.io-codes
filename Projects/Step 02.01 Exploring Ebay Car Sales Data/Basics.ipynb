## Initial Data Exploration and Cleaning
#### We will look for: - Text columns where all or almost all values are the same. These can often be dropped as they don't have useful information for analysis. - Examples of numeric data stored as text which can be cleaned and converted. 

```python
autos.describe(include='all')
```

###### Some columns seems to be identical (values are nearly all the same) : seller & offer_type
###### More investifations is needed for 'nr_of_pictures' as values are weird

```python
autos['nr_of_pictures'].value_counts()
```

##### nr_of_pictures column has 0 for every rows. Drop this column, plus the other two we noted as mostly one value¶

```python
autos = autos.drop(['nr_of_pictures', 'seller', 'offer_type'], axis=1)
```

#####  Change the price and odometer columns which are numeric values stored as text

```python
autos['price'] = (autos['price']
                 .str.replace('$', '')
                 .str.replace(',', '')
                 .astype(int) 
                 )
autos['price'].head(10)
```

```python
autos['odometer'] = (autos['odometer']
                     .str.replace(',', '')
                     .str.replace('km', '')
                     .astype(int)
                    )
autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
autos.head(8)
```

## Dealing with Incorrect Registration Year Data

##### A car can't be first registered before the listing was seen. Any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere around the 1945s. 

##### Percentage of our data with invalid values :

```python
(~autos['registration_year'].between(1945,2016)).sum() / autos.shape[0]
```

```python
autos = autos[autos['registration_year'].between(1945,2016)]
autos['registration_year'].describe()
```

```python
autos['registration_year'].value_counts(normalize=True).sort_index()
```

```python
autos['registration_year'].value_counts(normalize=True).sort_values(ascending=False).head(10)
```

## Exploration of price by brand

```python
autos['brand'].value_counts(normalize=True).head(10)
```

##### Volkswagen is by far the most popular brand, with approximately double the cars for sale of the next two brands combined. German manufacturers represent four out of the top five brands, almost 50% of the overall listings. 

```python
autos['brand'].value_counts(normalize=True, ascending=True).head(20)
```

##### More than the last 20 brands don't have a significant percentage:  our analysis will be limited to brands representing more than 1% of total listings (approximatively the first 20 brands).

```python
brand_counts = autos['brand'].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > 0.01].index
common_brands
```

```python
brand_mean_prices = {}

for brand in common_brands:
    #mean_price = autos[autos['brand']==brand, 'price'].mean()
    brand_name = autos[autos['brand'] == brand]
    mean_price = brand_name['price'].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

##### Most expensive manufacturers: audi, bmw, mercedes_benz  & skoda
##### Less expensive manufacturers: fiat, opel, peugeot, renault
##### In the top 5 most popular brands, there are both manufacturers with the highest and lowest mean prices. Further investigation towards the nb of damages (notRepairedDamage) or the mileage would be necessary to explain these results.

## Exploration of the mileage

```python
bmp_series = pd.Series(brand_mean_prices)
print(bmp_series)
```

```python
pd.DataFrame(bmp_series, columns=["mean_mileage"])
```

```python
brand_mean_mileage = {}

for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_mileage = brand_only['odometer_km'].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_mileage
```

```python
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending=False)
mean_prices
```

##### Among the most popular brands, all mileages are above 100 00km. On the contrary, there is a huge difference between the mean prices for each brand. -> The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.

```python
brand_info = pd.DataFrame(mean_mileage,columns=['mean_mileage'])
brand_info
```

```python
brand_info["mean_price"] = mean_prices
brand_info
```
